```python
import camelot
import camelot.io as camelot

# specify the PDF file to extract tables from
pdf_file = 'IMA_Master_List_(2023-03).pdf'

# create a list to store tables from all pages
all_tables = []

num_pages = 232
# iterate over the pages of the PDF
for page in range(3, num_pages+1):
    # The first two pages of the original document are an introduction, and the mineral list starts at three pages
    # extract tables from the current page
    tables = camelot.read_pdf(pdf_file, pages=str(page))
    # print(page, tables)
    print(page)
    # append the tables to the list
    all_tables.extend(tables)

# loop through each table and save as csv
# for i, table in enumerate(all_tables):
#    table.to_csv(f'/result/table_{i+1}.csv', index=False)
```

```python
import pandas as pd
dfs = [table.df for table in all_tables]

result = pd.concat(dfs)

# The variable result is the extracted information of all mineral lists
# result.to_csv("result/tables.csv", header=0, index=False)
```
