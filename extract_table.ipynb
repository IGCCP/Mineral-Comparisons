```python
import camelot
import camelot.io as camelot

tables = camelot.read_pdf('/content/drive/MyDrive/extractTable/IMA_Master_List_(2023-03).pdf')

table_df = tables[0].df

# Drop empty rows and columns
table_df.dropna(axis=0, how='all', inplace=True)
table_df.dropna(axis=1, how='all', inplace=True)

table_df.to_csv('table_data_camelot.csv', index=False, header=None)
```

```python
# specify the PDF file to extract tables from
pdf_file = '/content/drive/MyDrive/extractTable/IMA_Master_List_(2023-03).pdf'

# create a list to store tables from all pages
all_tables = []

num_pages = 230
# iterate over the pages of the PDF
for page in range(1, num_pages+1):
    # extract tables from the current page
    tables = camelot.read_pdf(pdf_file, pages=str(page))
    # print(page, tables)
    print(page)
    # append the tables to the list
    all_tables.extend(tables)

# loop through each table and save as csv
# for i, table in enumerate(all_tables):
#    table.to_csv(f'/content/drive/MyDrive/extractTable/result/table_{i+1}.csv', index=False)
```
